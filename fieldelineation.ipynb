```python
import rasterio as rio
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import skimage
import skimage.feature
```

### Sentinel 1.
[SENTINEL-1](https://directory.eoportal.org/web/eoportal/satellite-missions/c-missions/copernicus-sentinel-1)carries a single C-band synthetic aperture radar instrument operating at a centre frequency of 5.405 GHz. It includes a right-looking active phased array antenna providing fast scanning in elevation and azimuth, data storage capacity of 1 410 Gb and 520 Mbit/s X-band downlink capacity.


```python
#read_band Function reads tif files(tif files are the ) and retuns the raster data in ndarry format.  
def read_band(band,index):
    with rio.open(band) as src:
        band = src.read(index).astype('float64')
    return band

file1 = "./data/fd.tif" 
B1 = read_band(file1,1)
B2 = read_band(file1,2)
B3 = B2 -B1
```

```python

sigma = 5
low_threshold=0.10
high_threshold = 0.20
edges = skimage.feature.canny(
    image=B3,
    sigma=sigma,
    low_threshold=low_threshold,
    high_threshold=high_threshold,
)
plt.imshow(edges)
plt.axis('off')
plt.show()
```
